```python
# prompt: df에 있는 event_params의 한개의 내용을 보여줘

import pandas as pd

df = pd.read_csv('ga_innercircle.csv')

# Display the first event_params value
print(df['event_params'].iloc[0])
```

```python
import pandas as pd
import json

def parse_json_columns(df):
    # Identify columns that contain JSON-like structures
    json_columns = [col for col in df.columns if df[col].apply(lambda x: isinstance(x, str) and x.startswith('{')).any()]

    for col in json_columns:
        parsed_data = []

        for entry in df[col]:
            try:
                parsed_entry = {}
                if isinstance(entry, str):
                    json_obj = json.loads(entry)
                elif isinstance(entry, dict):
                    json_obj = entry
                else:
                    json_obj = {}

                if isinstance(json_obj, dict) and "event_params" in json_obj:
                    for param in json_obj.get("event_params", []):
                        key = param.get("key")
                        value_dict = param.get("value", {})

                        # Extract the first non-null value
                        value = next((v for v in value_dict.values() if v is not None), None)
                        if key:
                            parsed_entry[key] = value

                parsed_data.append(parsed_entry)
            except (json.JSONDecodeError, TypeError):
                parsed_data.append({})  # Append empty dict if parsing fails

        # Create new DataFrame from extracted JSON data
        new_df = pd.DataFrame(parsed_data)
        df = df.drop(columns=[col]).reset_index(drop=True)
        df = pd.concat([df, new_df], axis=1)

    return df


df = parse_json_columns(df)
df
```
